# Handling Imbalance (SMOTE for Oversampling)

```python
train_data.dropna()
train_data.drop_duplicates()
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

X = train_data.drop(columns=target_columns)
Y = train_data[target_columns]

Y = Y.applymap(lambda x: 1 if x > 0 else 0)

smote = SMOTE(random_state=42)

X_resampled_list = []
Y_resampled_list = []

for col in Y.columns:
    if len(Y[col].unique()) > 1:
        X_resampled, Y_resampled = smote.fit_resample(X, Y[col])
        X_resampled_list.append(X_resampled)
        Y_resampled_list.append(Y_resampled)
    else:
        print(f"Skipping SMOTE for column '{col}' as it has only one class.")

if X_resampled_list:
    balanced_train_data = X_resampled_list[0].copy()
    for i, col in enumerate(Y.columns):
        if col in balanced_train_data.columns:
            balanced_train_data[col] = Y_resampled_list[i]
        else:
            balanced_train_data[col] = Y[col]

    print("Dataset balanced using SMOTE (where applicable)!")
    print("New class distribution:\n", balanced_train_data[target_columns].sum())
else:
    print("No columns were resampled using SMOTE as all target columns had only one class.")
```

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

Y = Y.applymap(lambda x: 1 if x > 0 else 0)

X_under_list = []
Y_under_list = []

undersampler = RandomUnderSampler(random_state=42)

for col in Y.columns:
    if len(Y[col].unique()) > 1:
        X_resampled, Y_resampled = undersampler.fit_resample(X, Y[col])
        X_under_list.append(X_resampled)
        Y_under_list.append(Y_resampled)
    else:
        print(f"Skipping undersampling for column '{col}' as it has only one class.")
        X_under_list.append(X)
        Y_under_list.append(Y[col])

balanced_sp_dataset = pd.DataFrame(X_under_list[0])
for i, col in enumerate(Y.columns):
    balanced_sp_dataset[col] = Y_under_list[i]

print("Dataset balanced using Independent Undersampling for Each Label!")
print("New class distribution:\n", balanced_sp_dataset[target_columns].sum())
```

Import Libraries

```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
X = train_data.drop(["food_waste_kg"], axis=1)
```

```python
X_Scaled = Scaler.fit_transform(X)
Y = train_data["food_waste_kg"]
X_Train,X_Test,Y_Train,Y_Test=train_test_split(X_Scaled,Y,train_size=0.8,random_state=30)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_Train, Y_Train)

Y_Pred = model.predict(X_Test)

mae = mean_absolute_error(Y_Test, Y_Pred)
mse = mean_squared_error(Y_Test, Y_Pred)
rmse = mse ** 0.5
r2_score_value = r2_score(Y_Test, Y_Pred)

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R Score: {r2_score_value:.4f}")
```
